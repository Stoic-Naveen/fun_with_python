#Theoretical Questions


1.) What is a random variable in probability theory?

- A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. It's a way to map outcomes of a sample space to real numbers.


2.) What are the types of random variables?

- There are two main types:

- Discrete Random Variables: Can only take on a countable number of distinct values (e.g., number of coin flips, number of defective items).

- Continuous Random Variables: Can take on any value within a range or interval (e.g., height, temperature, time).


3.) What is the difference between discrete and continuous distributions?

- Discrete Distributions: Describe the probabilities of discrete random variables taking on specific values. They are often represented by probability mass functions (PMFs).

- Continuous Distributions: Describe the probabilities of continuous random variables falling within a range of values. They are represented by probability density functions (PDFs).


4.) What are probability distribution functions (PDF)?

- A PDF (for continuous random variables) is a function that describes the relative likelihood of a random variable taking on a given value. The area under the PDF curve within a certain interval represents the probability of the random variable falling within that interval.

5.) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

- PDF: Shows the probability density at each point (for continuous) or probability at each point (for discrete).

- CDF: Shows the probability that a random variable takes on a value less than or equal to a given value. It's the integral (for continuous) or sum (for discrete) of the PDF up to that point.


6.) What is a discrete uniform distribution?

- A discrete uniform distribution is a probability distribution where all possible values of a discrete random variable are equally likely.  Example: rolling a fair die.

7.) What are the key properties of a Bernoulli distribution?

- Represents a single trial with two possible outcomes (success or failure).
Has a single parameter 'p' representing the probability of success.
Mean = p, Variance = p(1-p)


8.) What is the binomial distribution, and how is it used in probability?

- The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials. It's used to model situations with a fixed number of trials, each with two possible outcomes and a constant probability of success.

9.) What is the Poisson distribution and where is it applied?

- The Poisson distribution models the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known average rate and independently of the time since the last event. It's used in areas like queuing theory, radioactive decay, and modeling website traffic.

10.) What is a continuous uniform distribution?

- A continuous uniform distribution is a probability distribution where all values within a specified interval are equally likely. Example: a random number generator that produces values between 0 and 1.

11.) What are the characteristics of a normal distribution?
- Bell-shaped and symmetrical.
- Defined by two parameters: mean (µ) and standard deviation (σ).
- Mean, median, and mode are equal.
- Approximately 68% of data falls within one standard deviation of the mean, 95% within two, and 99.7% within three.


12.) What is the standard normal distribution, and why is it important?
- The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. It's important because any normal distribution can be transformed into the standard normal distribution using z-scores, allowing for easy probability calculations.


13.) What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The CLT states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution. It's critical because it allows us to use normal distribution properties for statistical inference even when the population is not normally distributed.

14.) How does the Central Limit Theorem relate to the normal distribution?
- The CLT shows that the sampling distribution of the mean becomes approximately normal as the sample size grows, regardless of the population distribution. This allows us to use z-scores and normal distribution probabilities for hypothesis testing and confidence intervals.

15.) What is the application of Z statistics in hypothesis testing?
- Z-statistics are used to test hypotheses about population means when the population standard deviation is known or when the sample size is large enough for the sample standard deviation to be a good approximation.

16.) How do you calculate a Z-score, and what does it represent?
- The Z-score is calculated as:
- Z= (X-μ)/σ


- Where:

- X is the data point
- µ is the population mean
- σ is the population standard deviation It represents the number of standard deviations a data point is away from the mean.


17.) What are point estimates and interval estimates in statistics?
- Point Estimate: A single value used to estimate a population parameter (e.g., sample mean as an estimate of population mean).
- Interval Estimate: A range of values used to estimate a population parameter (e.g., confidence interval).


18.) What is the significance of confidence intervals in statistical analysis?
- Confidence intervals provide a range of plausible values for a population parameter, along with a level of confidence that the true parameter lies within that range. They help quantify the uncertainty associated with point estimates.

19.) What is the relationship between a Z-score and a confidence interval?
- Z-scores are used to determine the margin of error in a confidence interval. For a given confidence level (e.g., 95%), we find the corresponding Z-score (e.g., 1.96) and use it to calculate the interval.

20.) How are Z-scores used to compare different distributions?
- Z-scores standardize data points from different distributions to a common scale (standard normal distribution). This allows for meaningful comparisons of data points from different distributions.

21.) What are the assumptions for applying the Central Limit Theorem?
- Independence: The samples must be independent of each other.
- Sample Size: The sample size should be sufficiently large (generally n ≥ 30).
- Finite Variance: The population from which the samples are drawn must have a finite variance.


22.) What is the concept of expected value in a probability distribution?
- The expected value (or mean) of a random variable is the long-run average value of the variable. It's calculated as the sum (for discrete) or integral (for continuous) of each possible value multiplied by its probability.


23.) How does a probability distribution relate to the expected outcome of a random variable?
- The probability distribution provides the probabilities of all possible outcomes of a random variable. The expected value is the weighted average of these outcomes, where the weights are the probabilities. It represents the average outcome we expect in the long run.
