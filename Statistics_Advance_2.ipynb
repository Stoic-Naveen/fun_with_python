#Theoretical Questions

1.) What is hypothesis testing in statistics?

- Hypothesis testing is a statistical method used to make decisions or draw inferences about a population based on sample data. It's a formal process to check if there's enough evidence to reject a statement (null hypothesis) in favor of another statement (alternative hypothesis). We use sample data to assess the plausibility of a hypothesis about the population.


2.) What is the null hypothesis, and how does it differ from the alternative hypothesis?

- **Null Hypothesis (H₀) :** This is a statement of "no effect" or "no difference." It's the hypothesis we initially assume to be true and try to disprove. It often represents the status quo or a common belief. Examples: "There is no difference in average height between men and women," "The new drug has no effect."

- **Alternative Hypothesis (H₁ or Ha) :** This is the statement we are trying to find evidence for. It contradicts the null hypothesis. It often represents a new theory or an effect we suspect is present. Examples: "There is a difference in average height between men and women," "The new drug is effective."

- **Difference :** The null hypothesis is the hypothesis we test against, while the alternative hypothesis is what we accept if we reject the null hypothesis. They are mutually exclusive and exhaustive possibilities regarding the population.


3.) What is the significance level in hypothesis testing, and why is it important?

- **Significance Level (α) :** This is the probability of rejecting the null hypothesis when it is actually true (Type 1 error). It is set before conducting the hypothesis test and is usually a small value, commonly 0.05 (5%) or 0.01 (1%).

- **Importance :** The significance level sets the threshold for how much evidence we require to reject the null hypothesis. A smaller significance level (e.g., 0.01) makes it harder to reject the null hypothesis, reducing the risk of a Type 1 error but potentially increasing the risk of a Type 2 error. It determines the balance between these two types of errors and is crucial for making statistically sound decisions.


4.) What does a P-value represent in hypothesis testing?

- **P-value :** The P-value is the probability of observing sample data as extreme as, or more extreme than, the data actually observed, assuming the null hypothesis is true. It's calculated based on your sample data and the chosen test statistic.

- **What it represents :** A small P-value suggests that the observed data is unlikely to have occurred if the null hypothesis were true. It provides evidence against the null hypothesis. A large P-value suggests that the observed data is reasonably likely to have occurred under the null hypothesis, so we don't have enough evidence to reject it.


5.) How do you interpret the P-value in hypothesis testing?

- **Comparison with Significance Level (α) :** We compare the P-value to the significance level (α) to make a decision.

- If **P-value ≤ α :** We reject the null hypothesis (H₀). There is statistically significant evidence to support the alternative hypothesis.

- If **P-value > α :** We fail to reject the null hypothesis (H₀). There is not enough statistically significant evidence to support the alternative hypothesis. This does not mean we accept the null hypothesis as true, only that we don't have enough evidence to reject it.

- **Strength of Evidence :** A smaller P-value indicates stronger evidence against the null hypothesis.


6.) What are Type 1 and Type 2 errors in hypothesis testing?

- **Type 1 Error (False Positive) :** Rejecting the null hypothesis when it is actually true. We conclude there is an effect when there isn't one in the population. The probability of a Type 1 error is equal to the significance level (α).

- **Type 2 Error (False Negative) :** Failing to reject the null hypothesis when it is actually false. We fail to detect an effect that actually exists in the population. The probability of a Type 2 error is denoted by β.


7.) What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

- **Two-Tailed Test :** Used when the alternative hypothesis states that the population parameter is different from the value specified in the null hypothesis (it could be greater or smaller). We are looking for deviations in either direction from the null hypothesis value. The rejection region is split into two tails of the distribution.

- **One-Tailed Test :** Used when the alternative hypothesis states that the population parameter is either greater than or less than the value specified in the null hypothesis. We are only interested in deviations in one specific direction. The rejection region is in only one tail of the distribution.

- **Choosing the Tail :** The choice between one-tailed and two-tailed depends on the research question and the direction of the effect you are interested in detecting. Two-tailed tests are more common because they are less prone to bias if you are unsure of the direction of the effect.


8.) What is the Z-test, and when is it used in hypothesis testing?

- **Z-test :** A statistical test used to determine if there is a statistically significant difference between a sample mean and a known population mean, or between two sample means, when the population standard deviation is known, and the sample size is large enough (typically n ≥ 30) so that the Central Limit Theorem applies, or the population is normally distributed.

- **When to use :**
Population standard deviation (σ) is known.
Sample size is large (n ≥ 30) or the population is normally distributed.
Testing hypotheses about population means.


9.) How do you calculate the Z-score, and what does it represent in hypothesis testing?

- Z-score Formula (for a single sample mean compared to a population mean) :
Z = (x̄ - μ₀) / (σ / √n)

Where :

x̄ = sample mean

μ₀ = hypothesized population mean (from the null hypothesis)

σ = population standard deviation

n = sample size

**What it represents :** The Z-score represents how many standard deviations the sample mean (x̄) is away from the hypothesized population mean (μ₀) under the null hypothesis. It standardizes the sample mean difference into a scale that can be compared to the standard normal distribution. A large absolute Z-score indicates a large difference between the sample mean and the hypothesized population mean, suggesting evidence against the null hypothesis.


10.) What is the T-distribution, and when should it be used instead of the normal distribution?

- **T-distribution :** A probability distribution that is similar to the normal distribution but has heavier tails. It is used when the population standard deviation is unknown and estimated from the sample standard deviation. The shape of the t-distribution depends on the degrees of freedom (df), which are related to the sample size (usually df = n-1 for a one-sample t-test). As the degrees of freedom increase (larger sample size), the t-distribution approaches the normal distribution.

**When to use instead of normal distribution :**

- Population standard deviation (σ) is unknown and must be estimated from the sample (using sample standard deviation, s).

- Sample size is small (typically n < 30), and the population may not be normally distributed.

- The t-distribution accounts for the added uncertainty introduced by estimating the population standard deviation from the sample.


11.) What is the difference between a Z-test and a T-test?

- **Key Differences :**

- **Population Standard Deviation :** Z-test assumes population standard deviation (σ) is known; T-test is used when population standard deviation is unknown and estimated by sample standard deviation (s).

- **Sample Size :** Z-test is generally used for large samples (n ≥ 30) or when the population is known to be normal; T-test is more appropriate for small samples (n < 30) and is robust even if the population is not perfectly normal.
Distribution: Z-test uses the standard normal distribution; T-test uses the t-distribution.

- If we know the population standard deviation, or have a very large sample, we can use a Z-test. If we don't know the population standard deviation and have a smaller sample, use a T-test.


12.) What is the T-test, and how is it used in hypothesis testing?

- **T-test :** A statistical hypothesis test used to determine if there is a significant difference between the means of two groups (independent samples t-test) or between the means of paired observations (paired samples t-test), or to compare a sample mean to a hypothesized population mean (one-sample t-test) when the population standard deviation is unknown.

- How it's used in hypothesis testing :

- Formulate hypotheses (Null and Alternative).

- Choose a significance level (α).

- Calculate the t-statistic based on the sample data and the type of t-test (one-sample, independent samples, paired samples).

- The formula for the t-statistic will vary depending on the type of t-test. For a one-sample t-test comparing a sample mean to a hypothesized population mean (μ₀) :
t = (x̄ - μ₀) / (s / √n)
where 's' is the sample standard deviation.

- Determine the degrees of freedom (df).

- For a one-sample t-test, df = n - 1.

- For independent samples and paired samples t-tests, df calculation is slightly different.

- Find the P-value associated with the calculated t-statistic and degrees of freedom using the t-distribution.

- Compare the P-value to the significance level (α).

- If P-value ≤ α, reject the null hypothesis.

- If P-value > α, fail to reject the null hypothesis.

- Draw conclusions in the context of the problem.


13.) What is the relationship between Z-test and T-test in hypothesis testing?


- Both are hypothesis tests for means: Both Z-tests and T-tests are used to test hypotheses about population means.

- T-test as a generalization of Z-test: The t-test can be considered a more general test because it can be used when the population standard deviation is unknown (which is more common in real-world situations).

- **Convergence :** As the sample size increases, the t-distribution approaches the normal distribution. Therefore, for very large samples, the results of a Z-test and a T-test will be very similar. In practice, for very large samples, some people might use a Z-test even if the population standard deviation is unknown because the t-distribution is very close to the normal distribution.

- However, it's always safer and statistically more correct to use a t-test when the population standard deviation is unknown, regardless of sample size (although the difference becomes negligible with very large samples).

- **Key Distinction :** The crucial difference lies in whether the population standard deviation is known or unknown and the sample size.


14.) What is a confidence interval, and how is it used to interpret statistical results?

- **Confidence Interval (CI) :** A range of values, calculated from sample data, that is likely to contain the true population parameter with a certain level of confidence. It provides a plausible range for the true population parameter, rather than just a single point estimate. It's typically expressed as a percentage, e.g., a 95% confidence interval.

- **Interpretation :** A 95% confidence interval for the population mean, for example, means that if we were to repeat the sampling process many times and calculate a 95% confidence interval each time, approximately 95% of those intervals would contain the true population mean.

- **Use in Interpretation :**

- **Confidence intervals are used to :**

Estimate the range of plausible values for a population parameter.

- **Assess statistical significance :** If a confidence interval for the difference between two means does not contain zero, it suggests a statistically significant difference between the means (at the chosen confidence level).
If it does contain zero, it suggests no statistically significant difference.
- **Provide more information than just hypothesis tests :** Confidence intervals give you a range of plausible values and a sense of the precision of your estimate, which is often more informative than just a "reject" or "fail to reject" decision from a hypothesis test.


15.) What is the margin of error, and how does it affect the confidence interval?

- **Margin of Error (MOE) :** A measure of the uncertainty in a sample statistic as an estimate of a population parameter. It is the amount added and subtracted from the point estimate (e.g., sample mean) to create the confidence interval.

- **Relationship to Confidence Interval :**

- **The confidence interval is calculated as :**

- Confidence Interval = Point Estimate ± Margin of Error

For example, for a confidence interval for the population mean using a Z-statistic:

- Margin of Error = Z* * (σ / √n)
Where Z* is the critical Z-value corresponding to the desired confidence level.

- **How it affects CI :**

- **Larger MOE :** Wider confidence interval, indicating more uncertainty in the estimate.

- **Smaller MOE :** Narrower confidence interval, indicating more precision in the estimate.

- **Factors affecting MOE :**

- **Confidence Level :** Higher confidence level (e.g., 99% vs. 95%) leads to a larger margin of error (to be more confident, you need a wider interval).

- **Sample Size (n) :** Larger sample size leads to a smaller margin of error (more data reduces uncertainty).

- **Population Standard Deviation (σ) :** Larger population standard deviation leads to a larger margin of error (more variability in the population increases uncertainty).


16.) How is Bayes' Theorem used in statistics, and what is its significance?

- **Bayes' Theorem :** A fundamental theorem in probability theory that describes how to update the probability of a hypothesis based on new evidence.

It describes the probability of an event based on prior knowledge of conditions that might be related to the event. The formula is:

P(A|B) = [P(B|A) * P(A)] / P(B)

Where:

P(A|B): Posterior probability of hypothesis A given evidence B.

P(B|A): Likelihood of observing evidence B if hypothesis A is true.

P(A): Prior probability of hypothesis A (before observing evidence B).

P(B): Probability of observing evidence B (marginal likelihood or evidence).

- Use in Statistics :

- **Bayesian Statistics :** Bayes' Theorem is the cornerstone of Bayesian statistics, which provides a framework for updating beliefs in light of new data. It allows you to incorporate prior knowledge or beliefs (priors) with observed data (likelihood) to obtain updated probabilities (posteriors).

- **Updating Probabilities :** It's used to update probabilities when new information becomes available. For example, in medical diagnosis, you might start with a prior probability of a disease and then update this probability based on the results of a diagnostic test.

- **Machine Learning and AI :** Bayesian methods are widely used in machine learning, particularly in areas like Bayesian networks, spam filtering, and probabilistic modeling.

- **Significance :**

- **Incorporates Prior Knowledge :** Allows for the incorporation of prior beliefs or existing information, which can be valuable in many situations.

- **Probability of Hypotheses :** Focuses on the probability of hypotheses given the data, which is often what we are interested in (e.g., what is the probability that a patient has a disease given a positive test result?).

- **Adaptive Learning :** Provides a framework for learning and updating beliefs as more data becomes available.


17.) What is the Chi-square distribution, and when is it used?

- **Chi-square (χ²) Distribution :** A continuous probability distribution that arises frequently in statistics. It is defined by degrees of freedom (df). It is skewed to the right, and its shape depends on the degrees of freedom. It is used for tests involving categorical data.

- **When is it used :**

- **Chi-square Goodness of Fit Test :** To test if the observed frequencies of categorical data match expected frequencies from a hypothesized distribution.

- **Chi-square Test of Independence :** To test if two categorical variables are independent of each other.

- **Chi-square Test of Homogeneity :** To test if the distribution of a categorical variable is the same across different groups.

- **Variance Estimation :** In some cases, it's used in tests and confidence intervals related to variances of normally distributed populations.


18.) What is the Chi-square goodness of fit test, and how is it applied?

- **Chi-square Goodness of Fit Test :** A statistical test to determine if the observed frequencies of categorical data in a sample distribution match a hypothesized or expected distribution. It assesses how well the observed data "fits" the expected pattern.

- **Application Steps :**

- **State Hypotheses :**

- **Null Hypothesis (H₀) :** The observed frequencies match the expected frequencies (the data fits the hypothesized distribution).

- **Alternative Hypothesis (H₁) :** The observed frequencies do not match the expected frequencies (the data does not fit the hypothesized distribution).

- **Calculate Expected Frequencies (E) :** Based on the hypothesized distribution and the total sample size.

- Calculate the Chi-square test statistic (χ²) :

χ² = Σ [(Oᵢ - Eᵢ)² / Eᵢ]

Where:

Oᵢ = Observed frequency for category i.

Eᵢ = Expected frequency for category i.

Σ = Summation over all categories.

- **Determine Degrees of Freedom (df) :** df = (Number of categories) - 1

- **Find the P-value :** Using the chi-square distribution with the calculated df and the calculated χ² statistic.

- **Compare P-value to significance level (α) :**

- If P-value ≤ α, reject H₀. The data does not fit the hypothesized distribution.

- If P-value > α, fail to reject H₀. The data is consistent with the hypothesized distribution.

- **Draw conclusions :** Interpret the results in the context of the problem.


19.) What is the F-distribution, and when is it used in hypothesis testing?

- **F-distribution :** A continuous probability distribution that is skewed to the right and defined by two types of degrees of freedom: numerator degrees of freedom (df₁) and denominator degrees of freedom (df₂). It's used in tests involving ratios of variances.

- **When is it used in hypothesis testing :**

- **ANOVA (Analysis of Variance) tests :** The F-test is central to ANOVA for comparing means of two or more groups. ANOVA uses the F-distribution to test if there are significant differences in means by comparing variances between groups and within groups.

- **F-test for equality of variances :** To test if two populations have equal variances.

- **Regression analysis :** In regression, F-tests can be used to test the overall significance of the regression model.


20.) What is an ANOVA test, and what are its assumptions?

- **ANOVA (Analysis of Variance) :** A statistical test used to compare the means of two or more groups (levels of a categorical independent variable) to see if there is a statistically significant difference among them. It works by partitioning the total variance in the data into different sources of variation (between groups and within groups) and uses the F-test to assess if the between-group variance is significantly larger than the within-group variance, suggesting differences in means.

- **Assumptions of ANOVA :**

- **Independence :** Observations within and between groups are independent.

- **Normality :** The data for each group should be approximately normally distributed. (ANOVA is reasonably robust to violations of normality, especially with larger sample sizes).

- **Homogeneity of Variances (Homoscedasticity) :** The variances of the populations from which the samples are drawn are equal across all groups. This is often tested using Levene's test or Bartlett's test. Violation of this assumption can be problematic, especially with unequal sample sizes.


21.) What are the different types of ANOVA tests?

- **One-Way ANOVA :** Used when there is one categorical independent variable (factor) with two or more levels, and you want to compare the means of a continuous dependent variable across these levels. Example: Comparing the average exam scores of students taught using three different teaching methods.

- **Two-Way ANOVA :** Used when there are two categorical independent variables (factors), and you want to examine the effects of each factor and their interaction on a continuous dependent variable. Example: Studying the effects of both fertilizer type (Factor 1) and watering frequency (Factor 2) on plant growth. Two-way ANOVA can test for main effects of each factor and interaction effects between factors.

- **Repeated Measures ANOVA :** Used when the same subjects are measured multiple times under different conditions (or at different time points). It's used for within-subjects designs.

- **Example :** Measuring a patient's blood pressure at baseline, after one week of treatment, and after two weeks of treatment. Repeated measures ANOVA accounts for the correlation between measurements from the same subject.

- **MANOVA (Multivariate ANOVA) :** Used when there are multiple dependent variables, and you want to see if the independent variable(s) have an effect on a combination of these dependent variables.

- **Example :** Studying the effect of different exercise programs on both weight loss and cholesterol levels simultaneously.


22.) What is the F-test, and how does it relate to hypothesis testing?

- **F-test :** A statistical test that uses the F-distribution to determine if the ratio of two variances is significantly different from 1. In hypothesis testing, it is often used to compare variances or to test the overall significance of a model by comparing variances explained by the model to unexplained variances.

- **Relationship to Hypothesis Testing :**

- **ANOVA :** The F-test is the core component of ANOVA. In ANOVA, the F-statistic is calculated as the ratio of variance between groups to variance within groups (Mean Square Between / Mean Square Within). The null hypothesis in ANOVA is that all group means are equal. A large F-statistic (and small P-value) suggests that the variance between groups is large enough compared to the variance within groups to reject the null hypothesis, indicating that there are significant differences in group means.

- **Regression :** In regression analysis, an F-test can test the overall significance of the regression model. The null hypothesis is that all regression coefficients are zero (meaning the predictors don't explain variance in the dependent variable). The F-test in regression assesses if the model as a whole explains a significant portion of the variance in the dependent variable.
Equality of Variances: An F-test can be directly used to test if the variances of two populations are equal.

1.) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy import stats

def perform_z_test(sample_data, pop_mean, pop_std_dev, alpha=0.05, alternative='two-sided'):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = pop_std_dev / np.sqrt(sample_size)
    z_statistic = (sample_mean - pop_mean) / standard_error
    p_value = stats.norm.cdf(z_statistic)

    if alternative == 'two-sided':
        p_value = 2 * stats.norm.cdf(np.abs(z_statistic))
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)


    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest the sample mean is different from the population mean (p-value = {p_value:.4f} < alpha = {alpha})."
        if alternative == 'less':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest the sample mean is less than the population mean (p-value = {p_value:.4f} < alpha = {alpha})."
        elif alternative == 'greater':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest the sample mean is greater than the population mean (p-value = {p_value:.4f} < alpha = {alpha})."

    else:
        interpretation = f"Fail to reject the null hypothesis. There is not enough statistically significant evidence to suggest the sample mean is different from the population mean (p-value = {p_value:.4f} > alpha = {alpha})."

    return {
        "z_statistic": z_statistic,
        "p_value": p_value,
        "interpretation": interpretation
    }

sample_data = np.array([102, 98, 105, 95, 103, 101, 99, 104, 97, 100])
pop_mean = 100
pop_std_dev = 5

results = perform_z_test(sample_data, pop_mean, pop_std_dev)
print(results)
```

4.) Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_two_tailed_z_test(sample_data, pop_mean, pop_std_dev, alpha=0.05):
    results = perform_z_test(sample_data, pop_mean, pop_std_dev, alpha, alternative='two-sided')
    z_statistic = results["z_statistic"]
    p_value = results["p_value"]
    critical_z = stats.norm.ppf(1 - alpha / 2)
    lower_critical_z = -critical_z

    x = np.linspace(-4, 4, 200)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.axvline(z_statistic, color='red', linestyle='--', label=f'Z-statistic = {z_statistic:.2f}')
    plt.axvline(critical_z, color='green', linestyle='--', label=f'Critical Z = {critical_z:.2f}')
    plt.axvline(lower_critical_z, color='green', linestyle='--', label=f'Critical Z = {lower_critical_z:.2f}')

    x_fill_right = np.linspace(critical_z, 4, 100)
    y_fill_right = stats.norm.pdf(x_fill_right)
    plt.fill_between(x_fill_right, y_fill_right, color='lightcoral', alpha=0.5, label='Rejection Region')

    x_fill_left = np.linspace(-4, lower_critical_z, 100)
    y_fill_left = stats.norm.pdf(x_fill_left)
    plt.fill_between(x_fill_left, y_fill_left, color='lightcoral', alpha=0.5)


    plt.title('Two-Tailed Z-Test Visualization')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.text(0.5, 0.05, f'P-value = {p_value:.4f}', horizontalalignment='center', transform=plt.gca().transAxes)

    if p_value < alpha:
        plt.text(0.5, 0.02, 'Reject Null Hypothesis', horizontalalignment='center', color='red', transform=plt.gca().transAxes)
    else:
        plt.text(0.5, 0.02, 'Fail to Reject Null Hypothesis', horizontalalignment='center', color='green', transform=plt.gca().transAxes)

    plt.show()


sample_data = np.array([102, 98, 105, 95, 103, 101, 99, 104, 97, 100])
pop_mean = 100
pop_std_dev = 5
visualize_two_tailed_z_test(sample_data, pop_mean, pop_std_dev)
```

```python
import numpy as np
from scipy import stats

def compare_z_and_t_tests_simulation(sample_size, true_mean, hypothesized_mean, pop_std_dev, num_simulations=1000):
    z_p_values = []
    t_p_values = []

    for _ in range(num_simulations):
        simulated_data = np.random.normal(loc=true_mean, scale=pop_std_dev, size=sample_size)

        z_results = perform_z_test(simulated_data, hypothesized_mean, pop_std_dev, alternative='two-sided')
        z_p_values.append(z_results["p_value"])

        t_results = perform_one_sample_t_test(simulated_data, hypothesized_mean, alternative='two-sided')
        t_p_values.append(t_results["p_value"])

    avg_z_p_value = np.mean(z_p_values)
    avg_t_p_value = np.mean(t_p_values)

    print(f"--- Simulation Results ({num_simulations} simulations) ---")
    print(f"Sample Size: {sample_size}, True Mean: {true_mean}, Hypothesized Mean: {hypothesized_mean}, Pop Std Dev: {pop_std_dev}")
    print(f"Average P-value (Z-test): {avg_z_p_value:.4f}")
    print(f"Average P-value (T-test): {avg_t_p_value:.4f}")

def perform_one_sample_t_test(sample_data, hypothesized_mean, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample T-test. (Helper function for simulation)
    """
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)
    standard_error = sample_std_dev / np.sqrt(sample_size)
    t_statistic = (sample_mean - hypothesized_mean) / standard_error
    degrees_freedom = sample_size - 1
    p_value = stats.t.cdf(t_statistic, df=degrees_freedom)
    if alternative == 'two-sided':
        p_value = 2 * stats.t.cdf(np.abs(t_statistic), df=degrees_freedom)
    elif alternative == 'greater':
        p_value = 1 - stats.t.cdf(t_statistic, df=degrees_freedom)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest the sample mean is different from the population mean (p-value = {p_value:.4f} < alpha = {alpha})."
        if alternative == 'less':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest the sample mean is less than the population mean (p-value = {p_value:.4f} < alpha = {alpha})."
        elif alternative == 'greater':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest the sample mean is greater than the population mean (p-value = {p_value:.4f} < alpha = {alpha})."

    else:
        interpretation = f"Fail to reject the null hypothesis. There is not enough statistically significant evidence to suggest the sample mean is different from the population mean (p-value = {p_value:.4f} > alpha = {alpha})."


    return {
        "t_statistic": t_statistic,
        "p_value": p_value,
        "degrees_freedom": degrees_freedom,
        "interpretation": interpretation
    }

sample_size_small = 10
sample_size_large = 1000
true_mean = 52
hypothesized_mean = 50
pop_std_dev = 10

compare_z_and_t_tests_simulation(sample_size_small, true_mean, hypothesized_mean, pop_std_dev)
compare_z_and_t_tests_simulation(sample_size_large, true_mean, hypothesized_mean, pop_std_dev)
```

9.) Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_data, confidence_level=0.95, pop_std_dev=None):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    if pop_std_dev is not None:
        std_error = pop_std_dev / np.sqrt(sample_size)
        z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = z_critical * std_error
        interval = stats.norm.interval(confidence_level, loc=sample_mean, scale=std_error)
    else:
        sample_std_dev = np.std(sample_data, ddof=1)
        std_error = sample_std_dev / np.sqrt(sample_size)
        degrees_freedom = sample_size - 1
        t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_freedom)
        margin_of_error = t_critical * std_error
        interval = stats.t.interval(confidence_level, df=degrees_freedom, loc=sample_mean, scale=std_error)


    lower_bound, upper_bound = interval

    interpretation = f"We are {confidence_level*100:.0f}% confident that the true population mean lies within the interval ({lower_bound:.2f}, {upper_bound:.2f})."

    return {
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "margin_of_error": margin_of_error,
        "interpretation": interpretation
    }

sample_data = np.array([48, 52, 49, 51, 50, 53, 47, 50, 52, 49])
confidence_level = 0.95

pop_std_dev_known = 3
ci_results_z = calculate_confidence_interval(sample_data, confidence_level, pop_std_dev_known)
print("Confidence Interval (Z-interval, population std dev known):")
print(ci_results_z)

ci_results_t = calculate_confidence_interval(sample_data, confidence_level)
print("\nConfidence Interval (T-interval, population std dev unknown):")
print(ci_results_t)
```

13.) Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies_chi2(observed_frequencies):
    row_totals = observed_frequencies.sum(axis=1, keepdims=True)
    col_totals = observed_frequencies.sum(axis=0, keepdims=True)
    grand_total = observed_frequencies.sum()

    expected_frequencies = (row_totals @ col_totals) / grand_total

    return expected_frequencies

observed_data = np.array([[20, 30], [25, 15]])

expected_freqs = calculate_expected_frequencies_chi2(observed_data)
print("Observed Frequencies:\n", observed_data)
print("\nExpected Frequencies:\n", expected_freqs)
```

14.) Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

def perform_goodness_of_fit_test(observed_frequencies, expected_probabilities, alpha=0.05):
    n = np.sum(observed_frequencies)
    expected_frequencies = n * np.array(expected_probabilities)
    chi2_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)
    degrees_of_freedom = len(observed_frequencies) - 1

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis. The observed distribution does not fit the expected distribution (p-value = {p_value:.4f} < alpha = {alpha})."
    else:
        interpretation = f"Fail to reject the null hypothesis. The observed distribution is consistent with the expected distribution (p-value = {p_value:.4f} > alpha = {alpha})."

    return {
        "chi2_statistic": chi2_statistic,
        "p_value": p_value,
        "degrees_of_freedom": degrees_of_freedom,
        "expected_frequencies": expected_frequencies,
        "interpretation": interpretation
    }

observed_die_counts = np.array([12, 15, 9, 11, 14, 19])
expected_probabilities_uniform = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

results = perform_goodness_of_fit_test(observed_die_counts, expected_probabilities_uniform)
print(results)
```

16.) Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy import stats

def perform_f_test_variances(sample1_data, sample2_data, alpha=0.05, alternative='two-sided'):
    var1 = np.var(sample1_data, ddof=1)
    var2 = np.var(sample2_data, ddof=1)

    if var1 > var2:
        f_statistic = var1 / var2
        degrees_freedom_numerator = len(sample1_data) - 1
        degrees_freedom_denominator = len(sample2_data) - 1
    else:
        f_statistic = var2 / var1
        degrees_freedom_numerator = len(sample2_data) - 1
        degrees_freedom_denominator = len(sample1_data) - 1

    p_value = stats.f.cdf(f_statistic, degrees_freedom_numerator, degrees_freedom_denominator)

    if alternative == 'two-sided':
        p_value = 2 * min(p_value, 1 - p_value)
    elif alternative == 'greater':
        p_value = 1 - p_value

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis of equal variances. There is statistically significant evidence to suggest that the variances of the two populations are different (p-value = {p_value:.4f} < alpha = {alpha})."
        if alternative == 'less':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest that variance of sample 1 is less than variance of sample 2 (p-value = {p_value:.4f} < alpha = {alpha})."
        elif alternative == 'greater':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest that variance of sample 1 is greater than variance of sample 2 (p-value = {p_value:.4f} < alpha = {alpha})."
    else:
        interpretation = f"Fail to reject the null hypothesis of equal variances. There is not enough statistically significant evidence to suggest that the variances of the two populations are different (p-value = {p_value:.4f} > alpha = {alpha})."

    return {
        "f_statistic": f_statistic,
        "p_value": p_value,
        "degrees_freedom_numerator": degrees_freedom_numerator,
        "degrees_freedom_denominator": degrees_freedom_denominator,
        "interpretation": interpretation
    }

sample1 = np.array([25, 30, 28, 32, 29])
sample2 = np.array([22, 26, 24, 28, 25])
sample3 = np.array([10, 15, 8, 12, 9])

results_equal_var = perform_f_test_variances(sample1, sample2)
print("F-test for potentially equal variances:")
print(results_equal_var)

results_diff_var = perform_f_test_variances(sample1, sample3)
print("\nF-test for potentially different variances:")
print(results_diff_var)
```

17.) Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

```python
import numpy as np
from scipy import stats

def perform_anova_test(group_data_list, alpha=0.05):
    f_statistic, p_value = stats.f_oneway(*group_data_list)
    degrees_freedom_between = len(group_data_list) - 1
    degrees_freedom_within = sum([len(group) for group in group_data_list]) - len(group_data_list)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest that there is a difference in means among at least one of the groups (p-value = {p_value:.4f} < alpha = {alpha}). ANOVA does not tell you *which* groups are different, only that there is a difference somewhere."
    else:
        interpretation = f"Fail to reject the null hypothesis. There is not enough statistically significant evidence to suggest that there is a difference in means among the groups (p-value = {p_value:.4f} > alpha = {alpha})."

    return {
        "f_statistic": f_statistic,
        "p_value": p_value,
        "degrees_freedom_between": degrees_freedom_between,
        "degrees_freedom_within": degrees_freedom_within,
        "interpretation": interpretation
    }


group1 = np.array([25, 30, 28, 32, 29])
group2 = np.array([22, 26, 24, 28, 25])
group3 = np.array([35, 40, 38, 42, 39])

group_data = [group1, group2, group3]

results = perform_anova_test(group_data)
print(results)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def check_anova_assumptions(data, group_col, value_col):

    model = ols(f'{value_col} ~ C({group_col})', data=data).fit()

    residuals = model.resid
    shapiro_test = stats.shapiro(residuals)
    print(f"Shapiro-Wilk Test for Normality: W = {shapiro_test[0]}, p-value = {shapiro_test[1]}")

    stats.probplot(residuals, dist="norm", plot=plt)
    plt.title("Q-Q Plot of Residuals")
    plt.show()

    plt.scatter(model.fittedvalues, residuals)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel("Fitted Values")
    plt.ylabel("Residuals")
    plt.title("Residuals vs Fitted Values")
    plt.show()

    groups = [data[value_col][data[group_col] == group] for group in data[group_col].unique()]
    levene_test = stats.levene(*groups)
    print(f"Levene's Test for Equal Variances: W = {levene_test[0]}, p-value = {levene_test[1]}")

    bartlett_test = stats.bartlett(*groups)
    print(f"Bartlett's Test for Equal Variances: T = {bartlett_test[0]}, p-value = {bartlett_test[1]}")

data = pd.DataFrame({
    'group': np.repeat(['A', 'B', 'C'], 20),
    'value': np.concatenate([np.random.normal(10, 2, 20),
                             np.random.normal(12, 2, 20),
                             np.random.normal(15, 2, 20)])
})

check_anova_assumptions(data, 'group', 'value')
```

25.) Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from scipy import stats
import math

def perform_z_test_proportions(sample1_successes, sample1_size, sample2_successes, sample2_size, alpha=0.05, alternative='two-sided'):

    sample_proportion1 = sample1_successes / sample1_size
    sample_proportion2 = sample2_successes / sample2_size
    pooled_proportion = (sample1_successes + sample2_successes) / (sample1_size + sample2_size)

    standard_error = math.sqrt(pooled_proportion * (1 - pooled_proportion) * (1/sample1_size + 1/sample2_size))
    z_statistic = (sample_proportion1 - sample_proportion2) / standard_error

    p_value = stats.norm.cdf(z_statistic)

    if alternative == 'two-sided':
        p_value = 2 * stats.norm.cdf(np.abs(z_statistic))
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)

    interpretation = ""
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest that the population proportions are different (p-value = {p_value:.4f} < alpha = {alpha})."
        if alternative == 'less':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest that proportion 1 is less than proportion 2 (p-value = {p_value:.4f} < alpha = {alpha})."
        elif alternative == 'greater':
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence to suggest that proportion 1 is greater than proportion 2 (p-value = {p_value:.4f} < alpha = {alpha})."
    else:
        interpretation = f"Fail to reject the null hypothesis. There is not enough statistically significant evidence to suggest that the population proportions are different (p-value = {p_value:.4f} > alpha = {alpha})."

    return {
        "z_statistic": z_statistic,
        "p_value": p_value,
        "sample_proportion1": sample_proportion1,
        "sample_proportion2": sample_proportion2,
        "pooled_proportion": pooled_proportion,
        "interpretation": interpretation
    }

group1_successes = 60
group1_size = 100
group2_successes = 70
group2_size = 120

results = perform_z_test_proportions(group1_successes, group1_size, group2_successes, group2_size)
print(results)
```
